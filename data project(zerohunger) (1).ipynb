```python
#sample dada: years and hunger rates 
data={
    'year':[2010,2015,2016,2020],
    'goal_2_score':[10,40,20,40]
}
#creating a data frame 
df=pd.DataFrame(data)
#filter only the relavent columns for sdg 2
df_sdg2=df[['year','goal_2_score']]
#finding min value and corresponding year for sdg 2
min_sdg2=df_sdg2[df_sdg2['goal_2_score']==df_sdg2['goal_2_score'].min()]
#plotting the data 
plt.plot(min_sdg2['year'],min_sdg2['goal_2_score'],marker='o',markersize=10,color='r')
plt.text(min_sdg2['year'].values[0],min_sdg2['goal_2_score'].values[0],
         f'Minimum:{min_sdg2["year"].values[0]}',
         horizontalalignment='left',verticalalignment='bottom')
#adding labels and tittle
plt.xlabel('year')
plt.ylabel('SDG 2 Indicator')
plt.title('SDG 2 (zero hunger) over the years')
plt.grid(True)
plt.show()
```

```python

data = {
    'Year': list(range(2000, 2021)),
    'goal_2_score': [38.8, 58, 57, 55, 54, 52, 50, 49, 48, 47, 45, 43, 41, 40, 38, 36, 34, 32, 31, 30, 28]  
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['goal_2_score'], marker='o', linestyle='-', color='b', label='SDG 2 (Zero Hunger)')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('SDG 2 Indicator')
plt.title('SDG 2 (Zero Hunger) from 2000 to 2020')
plt.grid(True)
plt.xticks(df['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
#plotting graph in different year in different country
data = {
    'Year': [2000, 2000, 2000, 2010, 2010, 2010, 2020, 2020, 2020],
    'Country': ['Afghanistan', 'Albania', 'Angola', 'Afghanistan', 'Albania', 'Angola', 'Afghanistan', 'Albania ', 'Angola'],
    'goal_2_score': [50, 60, 55, 45, 65, 50, 40, 70, 45]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Filtering data for the years 2000 to 2020
df_filtered = df[(df['Year'] >= 2000) & (df['Year'] <= 2020)]

# Plotting the data
plt.figure(figsize=(12, 6))

# Grouping by year and country, then plotting
for year in df_filtered['Year'].unique():
    df_year = df_filtered[df_filtered['Year'] == year]
    plt.bar(df_year['Country'] + ' ' + df_year['Year'].astype(str), df_year['goal_2_score'], label=str(year))

# Adding labels and title
plt.xlabel('Country and Year')
plt.ylabel('SDG 2 Indicator')
plt.title('SDG 2 (Zero Hunger) for Multiple Countries (2000-2020)')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python

countries = ['Afghanistan', 'Albania', 'Angola', 'Bahrain', 'Colombia']
hunger_values = [50, 60, 70, 80, 90]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(hunger_values, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Hunger Values by Country for the Year 2022')
plt.show()
```
